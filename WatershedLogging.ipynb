### Calculate average level of disinfection byproducts

Data were downloaded for each water system from the Oregon Health Authority's online database: https://yourwater.oregon.gov/namelook.php. Data for individual systems were copied and pasted into one single file in Excel, by hand, because of formatting issues with the downloaded Excel files. Formatting issues relate to headers and do not affect the quality of the data.

---

```python
#read the file into pandas
dbps = pd.read_excel('data/dbp/dbp.xlsx')
dbps.head()
```

```python
# set '--' to null 
# then convert the columsn to floats so they can be averaged
dbps.loc[dbps['tthm'] == 'ND', 'tthm'] = '--'
dbps.loc[dbps['haa5'] == 'ND', 'haa5'] = '--'

dbps = dbps.replace('--',np.NaN)
dbps = dbps.replace('--',np.NaN)
dbps['tthm'] = dbps['tthm'].astype(float)
dbps['haa5'] = dbps['haa5'].astype(float)
```

```python
dbps_averages = pd.DataFrame({
        "tthm": dbps.groupby('system').mean().loc[:,'tthm'], 
        "haa5": dbps.groupby('system').mean().loc[:,'haa5']
    })
```
